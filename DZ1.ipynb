Кого больше - женщин или мужчин с ЗП >50K?

```python
df = pd.concat([X, y], axis=1)

male = len(df[(df['income'] == '>50K') & (df['sex'] == 'Male')])
female = len(df[(df['income'] == '>50K') & (df['sex'] == 'Female')])

print('Кого больше - женщин или мужчин с ЗП >50K?')
print(f'Male - {male}')
print(f'Female - {female}')
print(f'больше: {max(male, female)}')
```
